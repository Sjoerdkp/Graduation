```python
for year in range(2010, 2018):
    
    print(year)
    
    MON = pd.read_excel('OViN' + str(year) + '_Databestand.xlsx')
    
    # Save the number of respondents and movements
    totals.loc[year, 'Number of respondents'] = len(list(MON['OPID'].unique()))
    totals.loc[year, 'Number of movements'] = MON.shape[0]
          
    # Select the useful columns and rows (only the movements by car)
    MON = MON[['OPID', 'Jaardag', 'Weekdag', 'VertUur', 'AankUur', 'VertMin', 'AankMin', 'BVertUur', 'BVertMin', 'BAankUur', 'BAankMin', 'RVertUur', 'RVertMin', 'RAankUur', 'RAankMin', 'Rvm', 'Doel', 'Toer']]
    MON1 = MON
    MON1 = MON1[MON1['Rvm'] == 6]
    MON1 = MON1[MON1['Toer'] == 0]
    
    # For missing departure and arrival times of a specific movement: If the movement is for work, the departure and arrival times
    # can be found in a different column. Otherwise the departure and arrival time of the whole movement has been taken. If that 
    # also has missing values, the row has been dropped.
    MON1['RVertUur'] = MON1['RVertUur'].fillna(MON1['BVertUur'])
    MON1['RVertMin'] = MON1['RVertMin'].fillna(MON1['BVertMin'])
    MON1['RAankUur'] = MON1['RAankUur'].fillna(MON1['BAankUur'])
    MON1['RAankMin'] = MON1['RAankMin'].fillna(MON1['BAankMin'])
    
    MON1['RVertUur'] = MON1['RVertUur'].fillna(MON1['VertUur'])
    MON1['RVertMin'] = MON1['RVertMin'].fillna(MON1['VertMin'])
    MON1['RAankUur'] = MON1['RAankUur'].fillna(MON1['AankUur'])
    MON1['RAankMin'] = MON1['RAankMin'].fillna(MON1['AankMin'])
    
    MON1 = MON1.dropna(subset=['RVertUur'])
    MON1 = MON1.dropna(subset=['RAankUur'])
              
    # To make the rest of the code work, trips (partly) after midnight are being dropped. Also duplicates of trips by the 
    # same person at the same times are being dropped
    MON1 = MON1.drop_duplicates(subset = ['OPID', 'RVertUur', 'RAankUur'], keep = 'first')
    MON1 = MON1[MON1['RAankUur'] < 24]
    MON1 = MON1.reset_index(drop = True)
    
    ## Drop special days which dates differ per year:
    # Good Friday (02-04-2010, 22-04-2011, 06-04-2012, 29-03-2013, 18-04-2014, 03-04-2015, 25-03-2016, 14-04-2017)
    # Easter Monday (05-04-2010, 25-04-2011, 09-04-2012, 01-04-2013, 21-04-2014, 06-04-2015, 28-03-2016, 17-04-2017)
    # Queensday / Kingsday (30-04-2010, 30-04-2011, 30-04-2012, 30-04-2013, 26-04-2014, 26-04-2015, 26-04-2016, 26-04-2017)
    # Ascension Day (13-05-2010, 02-06-2011, 17-05-2012, 09-05-2013, 29-05-2014, 14-05-2015, 05-05-2016, 25-05-2017)
    # Friday after Ascension Day (14-05-2010, 03-06-2011, 18-05-2012, 10-05-2013, 30-05-2014, 15-05-2015, 06-05-2016, 26-05-2017)
    # Whit Monday (24-05-2010, 13-06-2011, 28-05-2012, 20-05-2013, 09-06-2014, 25-05-2015, 16-05-2016, 05-06-2017)
    for i in range(MON1.shape[0]):
        if MON1.loc[i, 'Jaardag'] in yeardict[str(year)]:
            MON1 = MON1.drop(i, axis = 0)
    MON1 = MON1.reset_index(drop = True)
    
    # Prepare dataframe in which the data of all days will be stored
    df_year = pd.DataFrame(index = [], columns = col5)
    
    for d in daylist:
        
        # Prepare new dataframe in which the data of a day will be stored
        df_day = pd.DataFrame(0, index = timelist, columns = col5)
        
        # Save number of respondents on the concerning day and the weekday by using the original data (MON instead of MON1)
        df_respondents = MON[MON['Jaardag'] == d]
        number_respondents = len(list(df_respondents['OPID'].unique()))
        
        L3 = [number_respondents] * 96
        df_driving = MON1[MON1['Jaardag'] == d]
        df_driving = df_driving.reset_index(drop = True)
        
        # Save Weekday (if df_driving is empty because the day has been dropped because it is a special day, the weekday has been calculated by using the last added day to df_year. For 1 January, the MON dataframe has been used)
        if df_driving.shape[0] > 1:
            df_day['Weekday'] = df_driving.loc[0, 'Weekdag']
        elif d == 1:
            MON_next_day = MON[MON['Jaardag'] == d + 1]
            MON_next_day = MON_next_day.reset_index(drop = True)
            if MON_next_day.loc[0, 'Weekdag'] > 1:
                df_day['Weekday'] = MON_next_day.loc[0, 'Weekdag'] - 1
            else:
                df_day['Weekday'] = 7
        else:
            if df_year.loc[df_year.shape[0] - 1, 'Weekday'] == 7:
                df_day['Weekday'] = 1
            else:
                df_day['Weekday'] = df_year.loc[df_year.shape[0] - 1, 'Weekday'] + 1
                   
        # Save a list of the driving respondents for iterating    
        personlist = list(df_driving['OPID'].unique())
        
        # Fill df_day with the data
        for p in personlist:
            save_personal_trips(p)
        
        # Add new columns to the dataframe df_day
        df_day['Number Respondents'] = L3
        df_day['Yearday'] = d
        df_day = df_day.reset_index(drop = True)
        df_day['Time'] = T15
        
        # Calculate number of cars at home, including cars that are at home for the full day
        df_day['Total Number Vehicles'] = df_day['Number Respondents'] * nr_cars / population # Number of cars that can be expected, given the size of the group of respondents
        df_day['Home'] = df_day['Total Number Vehicles']
        for c in col1[1:] + col2:
            df_day['Home'] = df_day['Home'] - df_day[c]
        
        # Convert to percentages
        for c in col5:
            df_day[c] = df_day[c] / df_day['Total Number Vehicles']  
        
        # Add the complete df_day to df_year to get a dataframe containing data of all the days
        df_year = df_year.append(df_day, ignore_index=True)

    # Save df_year
    df_year = df_year.reset_index(drop = True)      
    df_year.to_csv('Percentages yearday ' + str(year) + ' time step 15.csv')
    
totals.loc['Total', :] = totals.sum(axis = 0)
totals.to_csv('Number respondents and movements.csv')
```
