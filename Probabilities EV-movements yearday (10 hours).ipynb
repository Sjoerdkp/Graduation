```python
# Convert special days into nans
counter = 2010
for d in list_df:
    for i in range(0, EV_movements2010.shape[0], 96):
        if d.loc[i, 'Yearday'] in yeardict[str(counter)]:
            d.loc[i:i+95, 'Home':'Other_to_Driving_to_Personal_Care'] = 'nan'
    counter += 1
```

# Make years start on the same weekday

```python
special_days = [1, 125, 359, 360]
special_days_leap = [1, 126, 360, 361]

# Specify for every year how many days should be shifted
shiftdict = {
    '2011': 6,
    '2012': 5,
    '2013': 3,
    '2014': 2,
    '2015': 1,
    '2016': 0,
    '2017': 5
}
```

```python
# Calculate the averages at the 24 locations (1 hour)

for c in col6[3:27]:
    for t in range(EV_all_years.shape[0]):
        L = []
        L.append(EV_movements2010.loc[t, c])
        L.append(EV_movements2011.loc[t, c])
        L.append(EV_movements2012.loc[t, c])
        L.append(EV_movements2013.loc[t, c])
        L.append(EV_movements2014.loc[t, c])
        L.append(EV_movements2015.loc[t, c])
        L.append(EV_movements2016.loc[t, c])
        L.append(EV_movements2017.loc[t, c])
        L1 = [x for x in L if str(x) != 'nan']
        
        if len(L1) > 1:
            EV_all_years.loc[t, c] = (sum(L1) / len(L1))
```

```python
# Calculate the averages and standard deviations of the probabilities to start a trip (9 hours)

for c in col6[27:]:
    for t in range(EV_all_years.shape[0]):
        L = []
        L.append(EV_movements2010.loc[t, c])
        L.append(EV_movements2011.loc[t, c])
        L.append(EV_movements2012.loc[t, c])
        L.append(EV_movements2013.loc[t, c])
        L.append(EV_movements2014.loc[t, c])
        L.append(EV_movements2015.loc[t, c])
        L.append(EV_movements2016.loc[t, c])
        L.append(EV_movements2017.loc[t, c])
        L1 = [x for x in L if str(x) != 'nan']
        
        if len(L1) > 1:
            EV_all_years.loc[t, c + '_std'] = statistics.stdev(L1)
        else:
            EV_all_years.loc[t, c + '_std'] = 0
        if len(L1) > 0:
            EV_all_years.loc[t, c + '_av'] = (sum(L1) / len(L1))
        else:
            EV_all_years.loc[t, c + '_av'] = 0
```
